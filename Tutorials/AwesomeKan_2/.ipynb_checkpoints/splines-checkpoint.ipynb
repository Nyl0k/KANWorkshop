### What are B-Splines?

Splines are curves defined by a set of control points, where the curve smoothly follows said control points ("interpolation"). B-Splines ("basis splines") are splines that are C2-continuous, meaning their second derivative is continuous and well-defined everywhere. They are defined by two parameters:

**Order**

Order, or $k$, is the exponent of the interpolation curve. That is, what type of curve is used to move between control points. Some examples:

- $k=0$: The curve simply sits on the control point in question directly
- $k=1$: The curve linearly moves between two control points
- $k=2$: The curve quadtratically moves between three control points, and so on

Each curve is centered around a single control points, but is defined by the surrounding control points depending on the order, ultimately generating a smoother yet flatter and potentially less accurate curve depending on the order used.

**Control points**

Control points are defined points within the Euclidean mapping space (E-space) that the curve uses to define its path. More control points result in a more complex curve. The set of control points used is referred to as the "grid".

### Mathematical Framework

B-Splines are defined by a set of "basis functions" (hence the name basis spline), which generate "features". Each control point has a corresponding basis function. The basic representation of a basis function is:

$N_{i,k}(x): \mathcal{R} \rightarrow \mathcal{R}$

Where 

- $i$ is the number of the control point $(0,1,2...N)$
- $k$ is the order of the B-spline, as defined earlier
- $x$ is some arbitrary point along the x axis of the E-space

The calculation of the basis function is defined by the Cox-De Boor Recursion Formula. There are two primary cases for applying this formula. The first is when $k=0$. In that case, the basis function is defined as:

$N_{i,0}(x) =
    \begin{cases}
    1 & \text{if } t_i \leq x < t_{i+1} \\
    0 & \text{otherwise}
    \end{cases}
$

Where $t$ is the set of control points, and $t_i$ is the $i^{th}$ control point. This formula simply results in a one-hot encoding with a value of one from the control point to the next point along the spline, and zero everywhere else. Using this formula, if there are $M-1$ control points, there will be a total of $M$ features corresponding to a given $x$, because it simply has one output for every point for a given $x$, and each output generates a feature (strictly, each output *is* a feature).

The second case for applying the formula is if $k>0$. In that case, things become a little more complicated. The basis function is then defined as:

$N_{i,k}(x) = \frac{x - t_i}{t_{i+k} - t_i} N_{i,k-1}(x) + \frac{t_{i+k+1} - x}{t_{i+k+1} - t_{i+1}} N_{i+1,k-1}(x)$

While it may be messy to look at, essentially what this is doing is recursively taking the features generated by the basis functions of the next-lowest order for the current control point and the next control point in the sequence, multiplying them by the ratio of the distance from $x$ to the current control point over the distance from the next control point in the sequence to the current control point, and linearly combining the two values. What this does is generate an additional order of smoothness based on a point $k$ away in the sequence from the control point in question. By linearly combining and smoothing enough features from lower orders, complex and smooth curves can be generated as features at higher orders.

Or, in less complex terms, it takes the previous order's features for however many control points out the spline's order deals with, weights them based on the distance away, and combines them to generate higher-order features.

It is important to note that, because higher-order features require more control points, the number of possible features for $M$ control points becomes less. That is, only $M-k-1$ features can be generated for a B-Spline of order $k$ and with $M$ control points.


## A Program Example

The following code demonstrates the generation of B-splines via the Cox-De Boor Recursion Formula and confirms that their behavior is as predicted:

```python
#Here, we define the control points (using 10). 
grid = torch.linspace(-1, 1, steps=10)

#Define the span of x (the entire x-axis)
x = torch.linspace(-1, 1, steps=1000) 

#Reshape so that each x can be compared to each control point
grid_ = grid.unsqueeze(dim=0)
x_ = x.unsqueeze(dim=1)

#Cox-De Boor Recursion Formula base case (k=0)
k = 0
value1 = (x_ >= grid_[:,:-1]) * (x_ < grid_[:, 1:])

#Case for k=1
k = 1
value21 = (x_ - grid_[:, :-(k+1)]) / (grid_[:, 1:-k] - grid_[:, :-(k+1)]) * value1[:, :-1]
value22 = (grid_[:, (k+1):] - x_) / (grid_[:, (k+1):] - grid_[:, 1:-k]) * value1[:, 1:]
value2 = value21 + value22

#Case for k=2
k = 2
value31 = (x_ - grid_[:, :-(k+1)]) / (grid_[:, 1:-k] - grid_[:, :-(k+1)]) * value2[:, :-1]
value32 = (grid_[:, (k+1):] - x_) / (grid_[:, (k+1):] - grid_[:, 1:-k]) * value2[:, 1:]
value3 = value31 + value32

#Case for k=3
k = 3
value41 = (x_ - grid_[:, :-(k+1)]) / (grid_[:, 1:-k] - grid_[:, :-(k+1)]) * value3[:, :-1]
value42 = (grid_[:, (k+1):] - x_) / (grid_[:, (k+1):] - grid_[:, 1:-k]) * value3[:, 1:]
value4 = value41 + value42

#Confirm the number of basis functions, and therefore the number of features, is as expected
print(f"k=0, number of basis functions: {value1.shape[-1]}")
print(f"k=1, number of basis functions: {value2.shape[-1]}")
print(f"k=2, number of basis functions: {value3.shape[-1]}")
print(f"k=3, number of basis functions: {value4.shape[-1]}")
```

These numbers are consistent with theory of $M-k-1$ control points being available. Now let's plot the splines to confirm that the formula is working

```python
fig, axs = plt.subplots(ncols=2, nrows=2, figsize=(9, 9), dpi=100)

n_basis_to_plot = 4
all_basis = [(value1, 'k=0'), (value2, 'k=1'), (value3, 'k=2'), (value4, 'k=3')]
for i in range(4):
    
    ax = axs[i // 2, i % 2] 

    value, label = all_basis[i]
    for idx in range(value.shape[-1])[:n_basis_to_plot]:
        ax.scatter(x, value[:, idx], marker='x', label=f"Basis fn {idx}", alpha=0.5)
    ax.grid()
    ax.set_title(label)
    ax.legend()
```

These results are also as expected. The one-hot encoding triggers from each control point to the next (only 4 are shown), and the features are increasingly smooth and complex curves between those control points and the rest of the basis function, which is zero everywhere else. This is four curves rather than one because we are separating out the basis functions instead of making it one curve to better show how they work.

### Extending the Grid

The effect of less features being available at higher orders limits the representative power of B-splines at their extremes, because they have less points to interpolate between. To mitigate this effect, a technique called grid extension is used. Essentially, imaginary points are added to the grid that aren't a part of the function being approximated but follow what the function's path would likely be past the edge of its domain. The grid is extended by $k$ control points on each side to account for the loss of $k$ control points as order increases. We create a function to do this:

```python
def extend_grid(grid, k):
    """
    Extends the grid on either size by k steps

    Args:
        grid: number of splines x number of control points
        k: spline order

    Returns:
        new_grid: number of splines x (number of control points + 2 * k)
    """
    n_intervals = grid.shape[-1] - 1
    bucket_size = (grid[:, -1] - grid[:, 0]) / n_intervals
    
    for i in range(k):
        grid = torch.cat([grid[:, :1] - bucket_size, grid], dim=-1)
        grid = torch.cat([grid, grid[:, -1:] + bucket_size], dim=-1)

    return grid
```

Using that grid extension function, we will create a function to evalulate basis functions with the grid extensions.

```python
def eval_basis_functions(x_eval, grid, k):
    """
    Returns the value of basis functions defiend for order-k splines and control points defined in `grid`. 

    Args:
        x_eval: number of samples x number of dimensions
        grid: number of splines x number of control points
        k (scalar): order of spline

    Returns:
        bases: number of samples x number of dimensions x number of basis functions
    """
    grid_ = extend_grid(grid, k)
    # Reshape so that each x can be compared to each control point
    grid_ = grid_.unsqueeze(dim=2)
    x_ = x_eval.unsqueeze(dim=1)
    
    for idx in range(k+1):
        if idx == 0:
            bases = (x_ >= grid_[:, :-1]) * (x_ < grid_[:, 1:]) * 1.0 # step function; for each x in between the corresponding knots, the value is 1.
        else:
            bases1 = (x_ - grid_[:, :-(idx+1)]) / (grid_[:, 1:-idx] - grid_[:, :-(idx+1)]) * bases[:, :-1]
            bases2 = (grid_[:, (idx+1):] - x_) / (grid_[:, (idx+1):] - grid_[:, 1:-idx]) * bases[:, 1:]
            bases = bases1 + bases2

    return bases.transpose(1, 2) * 1.0
```

Finally, let's plot what the new basis functions look like now that grid extension has been implemented

```python
grid = torch.linspace(-1, 1, steps=10)
x = torch.linspace(-1, 1, steps=1000) # we take the entire domain to plot the basis function as a function of x

print(f"Grid: Shape:{grid.shape}, values:", grid)
print(f"x: Shape:{x.shape}, values:", x[:5])

# Reshape so that each x can be compared to each control point
grid_ = grid.unsqueeze(dim=0)
x_ = x.unsqueeze(dim=1)


value1 = eval_basis_functions(x_, grid_, k=0)

value2 = eval_basis_functions(x_, grid_, k=1)

value3 = eval_basis_functions(x_, grid_, k=2)

value4 = eval_basis_functions(x_, grid_, k=3)

print(f"k=0, number of basis functions: {value1.shape[-1]}")
print(f"k=1, number of basis functions: {value2.shape[-1]}")
print(f"k=2, number of basis functions: {value3.shape[-1]}")
print(f"k=3, number of basis functions: {value4.shape[-1]}")


fig, axs = plt.subplots(ncols=2, nrows=2, figsize=(9, 9), dpi=100)

n_basis_to_plot = 4
all_basis = [(value1, 'k=0'), (value2, 'k=1'), (value3, 'k=2'), (value4, 'k=3')]
for i in range(4):
    
    ax = axs[i // 2, i % 2] 

    value, label = all_basis[i]
    for idx in range(value.shape[-1])[:n_basis_to_plot]:  
        ax.scatter(x, value[:, 0, idx], marker='x', label=f"Basis fn {idx}", alpha=0.5)
    ax.grid()
    ax.set_title(label)
    ax.legend()
```

The shifting of the curves off the end of the grid due to the addition of imaginary points off the left side can clearly be seen

### Using B-Splines to Approximate Functions

Now that we have feature generation and grid extension working properly, we can use splines to emulate functions. The major motivation here is that B-splines are piecewise polynomials, which are very easy to work with mathematically, but can be used to very nearly approximate the curves of much more complex functions. We will define our approximation as $a(x)$, described by the formula

$a(x) = \sum_{i=0}^{n} c_i \cdot N_{i, k}(x) = \sum_{i=0}^{n} c_i \cdot N_i(x)$

To find the set of coefficients $c$ that best approximate some function $f(x)$, we use the operation $\min_{\mathbf{c}} |\mathbf{Nc} - f(x)|$

That is, find the value $c$ that minimizes the difference between the set of $Nc$ and $f(x)$

Let's do an example with code. We will use PyTorch's least squares function as the minimizer and attempt to approximate a sine wave:

```python
#Define the function to be approximated
fn = lambda x: torch.sin(torch.pi * x) # define a function that we need to match
#Find the coefficient set using least squares function
coeff1 = torch.linalg.lstsq(value1.squeeze(1), fn(x)).solution

#Multiply the coefficient vector with the feature vector of the 0th order B-spline to create the function approximation
y_pred = torch.einsum('i, ji -> j', coeff1, value1.squeeze(1))
plt.scatter(x, y_pred)
```

Pretty close, but it's clear that a 0th order approximation was being used with the feature sequence appearing as effectively a step function. Let's try a series of increasingly higher order approximations for a sine function and exponential sine function and see how it fares:

```python
fns = []
fns += [(lambda x: torch.sin(torch.pi * x), '$f(x) = sin(\pi*x)$')]
fns += [(lambda x: torch.sin(torch.pi * torch.exp(x)), '$f(x) = sin(\pi*e^{x})$')]

all_basis = [
        (eval_basis_functions(x_, grid_, k=0), 'k=0'), 
        (eval_basis_functions(x_, grid_, k=1), 'k=1'), 
        (eval_basis_functions(x_, grid_, k=2), 'k=2'), 
        (eval_basis_functions(x_, grid_, k=3), 'k=3')
]

fig, axs = plt.subplots(ncols=len(fns), nrows=len(all_basis), figsize=(15, 15), dpi=100)

for i in range(len(all_basis)):
    for j in range(len(fns)):
    
        ax = axs[i, j] 
        value, label = all_basis[i]
        fn, fn_label = fns[j]
        
        coeff = torch.linalg.lstsq(value.squeeze(1), fn(x)).solution # find the coefficients 
        y_pred = torch.einsum('i, ji -> j', coeff, value.squeeze(1)) # use these coefficients to evaluate y
        
        ax.scatter(x, fn(x), color='green', label='target')
        ax.scatter(x, y_pred, color='blue', label="predicted")

        ax.grid()
        ax.set_title(f"{label} - {fn_label}")
        ax.legend()
```

Much better. We should also showcase what happens if you don't use grid extension just as additional motivation for using it:

```python
grid = torch.linspace(-1, 1, steps=20) 
x = torch.linspace(-1, 1, steps=1000) # we take the entire domain to plot the basis function as a function of x

# Reshape so that each x can be compared to each control point
grid_ = grid.unsqueeze(dim=0)
x_ = x.unsqueeze(dim=1)

k = 0 # This gives us basis functions for order 0
value1 = (x_ >= grid_[:,:-1]) * (x_ < grid_[:, 1:])

k = 1 # This gives us basis functions for order 1
value21 = (x_ - grid_[:, :-(k+1)]) / (grid_[:, 1:-k] - grid_[:, :-(k+1)]) * value1[:, :-1]
value22 = (grid_[:, (k+1):] - x_) / (grid_[:, (k+1):] - grid_[:, 1:-k]) * value1[:, 1:]
value2 = value21 + value22

k = 2 # This gives us basis functions for order 2
value31 = (x_ - grid_[:, :-(k+1)]) / (grid_[:, 1:-k] - grid_[:, :-(k+1)]) * value2[:, :-1]
value32 = (grid_[:, (k+1):] - x_) / (grid_[:, (k+1):] - grid_[:, 1:-k]) * value2[:, 1:]
value3 = value31 + value32

k = 3 # This gives us basis functions for order 3
value41 = (x_ - grid_[:, :-(k+1)]) / (grid_[:, 1:-k] - grid_[:, :-(k+1)]) * value3[:, :-1]
value42 = (grid_[:, (k+1):] - x_) / (grid_[:, (k+1):] - grid_[:, 1:-k]) * value3[:, 1:]
value4 = value41 + value42


fns = []
fns += [(lambda x: torch.sin(torch.pi * x), '$f(x) = sin(\pi*x)$')]
fns += [(lambda x: torch.sin(torch.pi * torch.exp(x)), '$f(x) = sin(\pi*e^{x})$')]

all_basis = [(value1 * 1.0, 'k=0'), (value2, 'k=1'), (value3, 'k=2'), (value4, 'k=3')]

fig, axs = plt.subplots(ncols=len(fns), nrows=len(all_basis), figsize=(15, 15), dpi=100)

for i in range(len(all_basis)):
    for j in range(len(fns)):
    
        ax = axs[i, j] 
        value, label = all_basis[i]
        fn, fn_label = fns[j]
        
        coeff = torch.linalg.lstsq(value, fn(x)).solution # find the coefficients 
        y_pred = torch.einsum('i, ji -> j', coeff, value) # use these coefficients to evaluate y
        
        ax.scatter(x, fn(x), color='green', label='target')
        ax.scatter(x, y_pred, color='blue', label="predicted")

        ax.grid()
        ax.set_title(f"{label} - {fn_label}")
        ax.legend()
```

As you can see, the B-spline seriously loses approximation power near the edges. Finally, an example with noise:

```python
fns = []
fns += [(lambda x: torch.sin(torch.pi * x) + 0.1 * torch.randn(x.shape), '$f(x) = sin(\pi*x) + \epsilon$')]
fns += [(lambda x: torch.sin(torch.pi * torch.exp(x)) + 0.1 * torch.randn(x.shape), '$f(x) = sin(\pi*e^{x}) + \epsilon$')]

all_basis = [
        (eval_basis_functions(x_, grid_, k=0), 'k=0'), 
        (eval_basis_functions(x_, grid_, k=1), 'k=1'), 
        (eval_basis_functions(x_, grid_, k=2), 'k=2'), 
        (eval_basis_functions(x_, grid_, k=3), 'k=3')
]

fig, axs = plt.subplots(ncols=len(fns), nrows=len(all_basis), figsize=(15, 15), dpi=100)

for i in range(len(all_basis)):
    for j in range(len(fns)):
    
        ax = axs[i, j] 
        value, label = all_basis[i]
        fn, fn_label = fns[j]

        y_target = fn(x)
        coeff = torch.linalg.lstsq(value.squeeze(1), y_target).solution # find the coefficients 
        y_pred = torch.einsum('i, ji -> j', coeff, value.squeeze(1)) # use these coefficients to evaluate y
        
        ax.scatter(x, y_target, color='green', label='target')
        ax.scatter(x, y_pred, color='blue', label="predicted")

        ax.grid()
        ax.set_title(f"{label} - {fn_label}")
        ax.legend()
```
